```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of the target variable (diagnosis of heart disease)
plt.figure(figsize=(8, 6))
sns.countplot(x='num', data=y, palette='viridis', hue='num', legend=False)
plt.title('Distribution of Heart Disease Diagnosis')
plt.xlabel('Diagnosis of Heart Disease (0 = no presence, 1-4 = presence)')
plt.ylabel('Number of Patients')
plt.xticks(ticks=[0, 1, 2, 3, 4], labels=['No Disease', 'Stage 1', 'Stage 2', 'Stage 3', 'Stage 4'])
plt.show()
```

### Dataset Information: Heart Disease

Based on the fetched metadata and variable information, here are the details about this dataset:

*   **Source/Repository URL:** https://archive.ics.uci.edu/dataset/45/heart+disease
*   **Dataset Name:** Heart Disease
*   **Abstract:** This dataset comprises 4 databases: Cleveland, Hungary, Switzerland, and the VA Long Beach.
*   **Creators:** Andras Janosi, William Steinbrunn, Matthias Pfisterer, Robert Detrano
*   **Year of Dataset Creation:** 1989
*   **Last Updated:** Fri Nov 03 2023
*   **Tasks:** Classification
*   **Characteristics:** Multivariate
*   **Number of Instances:** 303
*   **Number of Features:** 13
*   **Feature Types:** Categorical, Integer, Real
*   **Demographics:** Includes Age and Sex information.
*   **Target Column:** 'num' (diagnosis of heart disease)
*   **Missing Values:** Yes, indicated by 'NaN'.

#### Additional Important Information from `additional_info`:

*   **Summary:** This database originally contained 76 attributes, but published experiments typically use a subset of 14. The Cleveland database is the most commonly used by ML researchers. The 'goal' field refers to the presence of heart disease (values 1,2,3,4) vs. absence (value 0).
*   **Variable Information:** Only 14 attributes were used in published experiments, including:
    *   `age`: age in years
    *   `sex`: sex (1 = male; 0 = female)
    *   `cp`: chest pain type
    *   `trestbps`: resting blood pressure (in mm Hg)
    *   `chol`: serum cholestoral in mg/dl
    *   `fbs`: fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
    *   `restecg`: resting electrocardiographic results
    *   `thalach`: maximum heart rate achieved
    *   `exang`: exercise induced angina (1 = yes; 0 = no)
    *   `oldpeak`: ST depression induced by exercise relative to rest
    *   `slope`: the slope of the peak exercise ST segment
    *   `ca`: number of major vessels (0-3) colored by flourosopy
    *   `thal`: 3 = normal; 6 = fixed defect; 7 = reversable defect
    *   `num`: diagnosis of heart disease (0 = < 50% narrowing, 1 = > 50% narrowing)
